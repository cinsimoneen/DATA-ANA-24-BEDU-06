```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

FILENAME = 'tmp/Ejemplo_02_03_Cybersecurity_Dataset.csv'

# Cargar nuestro dataset

df = pd.read_csv(FILENAME)
df.head()

# Configuracion inicial para seaborn
sns.set(style='darkgrid')

# Configurar scatter
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='severity', y='response_time', color='b', s=100)

# Configurar titulo y etiquetas
plt.title('Severidad VS Tiempo de respuesta en ataques de ciber seguridad')
plt.xlabel('Severidad del incidente')
plt.ylabel('Tiempo de respuesta')

# mostrar grafico
plt.show()
```
